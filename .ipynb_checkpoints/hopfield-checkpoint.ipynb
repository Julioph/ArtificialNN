```python
def create_patterns(n):
    bits = 120
    P = np.empty(shape=(n,bits))
    for i in range(n):
        U = np.random.randint(0,2,bits)
        for r in range(len(U)):
            if U[r] == 0:
                U[r] = -1
                
        P[i] = U
        
    return P

def create_W(X):
    W = np.outer(np.transpose(X), X) - np.identity(len(X))
    
    return W

def _store_patterns(patterns_array):
    num_patterns = 0
    for p in patterns_array:
        if num_patterns == 0:
            W = create_W(p)
            num_patterns += 1 
        else:
            new = create_W(p)
            W = W + new
            num_patterns += 1
    return W/160

def store_patterns(matrix):
    # X = np.matmul(np.matrix.transpose(matrix), matrix)
    # Y = X.shape[0]*np.identity(n=len(X))
    # return (X - Y)/10
    X =  np.outer(matrix, np.matrix.transpose(matrix))/160
    return X
    

def update(weights, in_vector, times=1):
    error = 0
    for s in range(times):
        neuron_i = np.random.randint(0,len(in_vector))
        out = np.dot(weights[neuron_i][:], in_vector)
        temp = np.copy(in_vector)
        if out >= 0:
            in_vector[neuron_i] = 1
        else:
            in_vector[neuron_i] = -1 #can simplify
            
        comparison = temp == in_vector
        if comparison.all():
        # No error, in_vec is same as stored
            pass
        else:
        # Error. in_vec is not same as stored
            error += 1
            

    return in_vector, error

def update_async(weights, in_vector, t):
#     neuron_i = np.random.randint(0,in_vector.shape[0])
    for neuron_i in range(weights.shape[0]):
        out = np.dot(weights[neuron_i][:], in_vector)
        temp = np.copy(in_vector)    
        
        if out >= 0:
            in_vector[neuron_i] = 1
        else:
            in_vector[neuron_i] = -1 #can simplify
            
        comparison = temp == in_vector
        if comparison.all():
        # No error, in_vec is same as stored
            pass
        else:
        # Error. in_vec is not same as stored
            print("babokio!", neuron_i, out, t)
    

            

    return in_vector
```
